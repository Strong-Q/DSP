## Problem 3.

Run the following code which explains how to make a kind of animation effect in Matlab.

```octave
%1)

figure(1); hold on
axis([-1 1 -4 .5])

for x = -1:0.05:1
    y = -4*x^2;
    plot(x,y,'s')
    pause(0.1)
end
hold off
```

## Problem 4.

Write a code to create a circle as shown in the below figure. (hint: there are many ways to realize this problem, but think about Euler’s formula (http://en.wikipedia.org/wiki/Euler's_formula))  

```octave
y = -1:0.01:1;
x1 = sqrt(1 - y.^2);
x2 = -sqrt(1 - y.^2);

plot(y,x1), hold on
plot(y,x2), hold off
axis equal
```

```octave
[x,y] = scircle1(0,0,1);
plot(x,y)
axis equal
```

```octave
center = [0 0];                 % 중심점
r = 70;                         % 반지름
N = 1000;                       % 원의 점 개수
NL = 600;                       % 선의 점 개수*2
theta = linspace(0,2*pi,N);     % 원(라디안)
x = r*cos(theta) + center(1);       
y = r*sin(theta) + center(2);       
plot(x,y);                      

axis equal;                     % 좌표축.
grid on
```

```octave
t = [0:0.1:2*pi];

% 키보드로부터 입력한다

r  = 1;
x0 = 0;
y0 = 0;

% 시간 t=0~2pi 까지 돌리면서 원의 x좌표와 y좌표를 뽑아낸다.=> 반복문 사용: for문

for j = 1:length(t),
    x(j) = x0 + r*cos(t(j));
    y(j) = y0 + r*sin(t(j));
end

% 그래프를 그린다

plot(x,y)
title('Circle')
xlabel('x(t)')
ylabel('y(t)')
```

## Problem 5.

Create an animation which the green dot is moving around the unit circle (counter clockwise). 

```octave
[x,y] = scircle1(0,0,1);

for y1 = -1:0.01:1;
    x1 = sqrt(1 - y1.^2);
    plot(x1,y1,'s',x,y)
    pause(0.01)
end

for y2 = -1:0.01:1;
    x2 = -sqrt(1 - y2.^2);
    plot(x2,-y2,'s',x,y)
    pause(0.01)
end
```
