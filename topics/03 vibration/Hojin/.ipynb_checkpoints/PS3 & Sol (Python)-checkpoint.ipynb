## Problem 1.

Here, weâ€™d like to numerically find the solutions (roots) of the following equation,

$$f(x) = 4x^3 + 10x^2 - 20x - 10 \quad \text{for} \quad -4 \leq x \leq 3 $$

1) Plot $f(x)$ over $-4 \leq x \leq 3$

2) Numerically find all the roots (solutions) of $f(x) = 0$ with a hint of the above figure.

```python
# 1)

x = np.arange(-4.0,3.0,0.01)
f = 4*x**3 + 10*x**2 - 20*x - 10

plt.plot(x,f)
plt.show()
```

```python
# 2)

f2 = [4.0,10.0,-20.0,-10.0]
x = np.roots(f2)

print(x)
```

## Problem 4.

Write a code to create a circle as shown in the below figure. (hint: there are many ways to realize this problem, but think about Eulerâ€™s formula (http://en.wikipedia.org/wiki/Euler's_formula))  

```python
y = np.arange(-1.0,1.00,0.01)

x1 = np.sqrt(1-y**2)
x2 = -np.sqrt(1-y**2)

plt.plot(y,x1)
plt.plot(y,x2)
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
center = [0.0,0.0]

r = 70.0

N = 1000
NL = 600

theta = np.linspace(0.0,2*np.pi,N)

x = r*np.cos(theta) + center[0]
y = r*np.sin(theta) + center[1]

plt.plot(x,y)
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
t = np.arange(0.0,2*np.pi,0.1)

r = 1
x0 = 0
y0 = 0
x = []
y = []

for j in range(0,len(t)):
    x.append(x0+r*np.cos(t[j]))
    y.append(y0+r*np.sin(t[j]))
    
plt.plot(x,y)
plt.axis('equal')
plt.grid(True)
plt.title('Circle')
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.show()
```
